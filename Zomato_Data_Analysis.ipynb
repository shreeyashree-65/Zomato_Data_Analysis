```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts()
```

Con: Max transaction is in India

```python
final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_val=final_df.Country.value_counts().values
```

**Creating Pie Chart**

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%') #To check which country has maximum transaction or orders (top 3)
```

Observation: Maximum records of Zomato is in India followed by United States and United Kingdoms

```python
from typing_extensions import final
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

Con: Max number of ratings is 0 i.e, many users have not rated the app yet

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) #To zoom the graph
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color', data=ratings)
```

Mapping the colors

```python
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color',data=ratings, palette=['white','red', 'yellow', 'orange', 'green', 'green'])
```

Con:
  1. Not rated count is very high
  2. Maximum number of ratings are between 2.5 to 3.4
  

```python
#Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red', 'yellow', 'orange', 'green', 'green'])
```
